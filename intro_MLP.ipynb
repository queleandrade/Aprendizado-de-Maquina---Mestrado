```python
import numpy as np
import matplotlib.pyplot as plt

from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense

from tensorflow.keras.utils import to_categorical
```

```python
# рҹ“Ң Carregar dados
X, y = load_iris(return_X_y=True)
```

```python
y
```

```python
y = to_categorical(y)
```

```python
y
```

```python
X.shape
```

```python

# рҹ“Ң DivisГЈo treino/teste
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
X_train.shape
```

```python
X_train
```

```python

# рҹ“Ң Modelo MLP
model = Sequential([
    Dense(8, activation='relu', input_shape=(4,)),
    Dense(3, activation='softmax')
])
```

```python
model.summary()
```

```python

# рҹ“Ң CompilaГ§ГЈo
model.compile(optimizer='adam',
              loss='categorical_crossentropy',
              metrics=['accuracy'])

# рҹ“Ң Treinamento com validaГ§ГЈo
history = model.fit(
    X_train, y_train,
    validation_data=(X_test, y_test),
    epochs=100,
    batch_size = 16,
    verbose=1
)
```

```python
import numpy as np
import matplotlib.pyplot as plt

# рҹ“Ҡ ===============================
# рҹ“ү PLOT DO LOSS
# рҹ“Ҡ ===============================
plt.figure()
plt.plot(history.history['loss'], label='Treino')
plt.plot(history.history['val_loss'], label='ValidaГ§ГЈo')
plt.title('Loss ao longo das Г©pocas')
plt.xlabel('Гүpocas')
plt.ylabel('Loss')
plt.legend()
plt.show()

# рҹ“Ҡ ===============================
# рҹ“Ҳ PLOT DA ACURГҒCIA
# рҹ“Ҡ ===============================
plt.figure()
plt.plot(history.history['accuracy'], label='Treino')
plt.plot(history.history['val_accuracy'], label='ValidaГ§ГЈo')
plt.title('AcurГЎcia ao longo das Г©pocas')
plt.xlabel('Гүpocas')
plt.ylabel('AcurГЎcia')
plt.legend()
plt.show()
```
